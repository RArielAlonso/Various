```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('test_database') 
c = conn.cursor()
                 
c.execute('''
          SELECT
          *
          FROM products
          ''')

Tabla = pd.DataFrame(c.fetchall(), columns = ['product_id', 'product_name', 'price'])
print (Tabla)
```
